### Definicin y producto punto

Un producto interno o producto interior, sobre un espacio vectorial (para el caso de este notebook ser $R^n$), es una operacion que asigna a cada par de vectores u y v un escalar tal que <u|v> y cumple con las siguinetes propiedades:

* <v|v> $\geq 0$
* <v|v> = 0 si y solo si v =0
* <u|v+w> = <u|v> + <u|w>
* Sea $\alpha \in R$ entonces < $\alpha$ v|w> = $\alpha$<v|w>
* <v|w>= $\bar{<w|v>}$ (Esta ltima propiedad se usa con los nmeros complejos, por lo que para fines de este notebook no se usar pero es importante saber de su existencia)

En el caso de $ R^n$ el porducto interior es mejor conocido como producto punto y est definido como $\vec{a} \cdot \vec{b} = ||\vec{a}||||\vec{b}||cos $ donde $||\vec{a}||$ y $||\vec{b}||$ son la magnitud del vector $\vec{a}$ y $\vec{b}$ respectivamente y $$ el ngulo entre ambos.



Ten en cuenta que $\textbf{el producto punto de dos vectores es un escalar, no un vector.}$ Eso significa, por ejemplo, que no tiene sentido preguntar a qu es igual $\vec{a} \cdot \vec{b} \cdot \vec{c} $.

Una vez que hayamos evaluado $\vec{a} \cdot \vec{b}$
 y obtenido un nmero, terminaramos por tratar de tomar el producto punto entre un nmero y un vector.

 Otra forma equivalente y ms sencilla de calcular el producto punto es conociendo los coeficientes de sus componentes; de modo que el produto punto ser la suma de las multiplicaciones de sus componentes.

Supongamos que tenemos a los vectores 

$x = (x_1,x_2,...,x_n)$

$y = (y_1,y_2,...,y_n) $


Tenemos que el producto interior entonces ser:

$\sum_{i=1}^n x_iy_i$

Ahora para visualizar mejor el porducto punto vemos un ejemplo.

## Ejemplo de producto punto en $R^n$
Si tenemos los vectores 
$x = (3,2)$
$y = (5,6)$
 Entonces el producto punto ser $\alpha = (3)(5)+(2)(6)= 20$

## Una ballena que cae del cielo

Supongamos que una ballena llamada Whilly, cae del cielo, y que cae a lo largo de una trayectoria curvada, porque las corrientes de aire la empujan de un lado para otro.



<img src = 'Whilly.svg'>
<figcaption> <center> Figura 1 Whilly y su trayectoria curva.  </center> </figcaption>

En este ejemplo, se asume que el alumno est familiarizado con las ideas de la fsica que la fuerza hace trabajo sobre un objeto en movimiento, adems la fuerza est definida como 
$$
\begin{align}
{\vec{W} = \vec{F} \cdot \vec{s}}
\end{align}
$$

es decir que el trabajo est definido como el producto punto entre el vector de fuerza y el vector de desplazamiento.

$\textbf{Pregunta clave:}$ cul es el trabajo que la fuerza de gravedad realiza sobre Whilly conforme cae a lo largo de la trayectoria curvada C?


Usualmente, calculamos el trabajo con respecto a un vector constante de fuerza y un vector constante de desplazamiento. Qu podemos hacer con una trayectoria curvada? 

Podemos comenzar imaginndonos que la curva est compuesta de muchos vectores pequeos de desplazamiento:

$\vec{\Delta s_1}, \vec{s_2}, ...$

El trabajo realizado por la gravedad a lo largo de cada uno de estos vectores de desplazamiento es el producto punto del vector de fuerza de gravedad, que denotaremos $\vec{F_g} $, con el vector de desplazamiento mismo por lo que la contibcin de los vectores se ver como:

$_i^n\vec{F_g} \cdot \vec{s_i}$




```python
import matplotlib.pyplot as plt #Se importa la paquetera que genera grficas
import numpy as np #Se importa la paquetera que tiene acceso a la parte matemtica
import ipywidgets as widgets #Se importa la paquetera para la generacin de la parte interactiva con el usuario



# make data
y = np.linspace(0, 10, 200)
x = 4 + 2 * np.sin(2 * x)

# plot
fig, ax = plt.subplots()

ax.plot(x, y+x, linewidth=2.0)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

Pero lo que realmente nos interesa es el valor lmite de esta suma conforme el tamao de esos pasos es ms y ms pequeo. Para ello haremos uso de la integral ya que toma en cuenta las contibuciones de las trayectorias infinitesimales y se ver de la forma:

$_C \vec{F}\cdot \vec{ds}$
Similar a una integral de lnea en un campo escalar pero ahora consideramos el pequeo paso $\vec{ds} $ como un vector, no como una longitud.


$\textbf{Nota}$: la cantidad que sumamos conforme caminamos a lo largo de C no es todo el valor de $\vec{F_g}$ en cada punto, sino la componente de $\vec{F_g}$ que apunta en la misma direccin que el vector $\vec{ds}$. Es decir, la componente de la fuerza en la direccin de la curva.
